```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
df = pd.read_csv(Path('../data/BTC-USD_features.csv'), parse_dates=['Date'])
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Closing Price Trend analysis
plt.figure(figsize=(8,5))
plt.plot(df['Date'], df['Close'], label="Close Price")
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.title('BTC-USD Closing Price Over Time')
plt.legend()
plt.show()
```

```python
#Volatility Bands
plt.figure(figsize=(8,5))
plt.plot(df['Date'], df['Close'], label="Close Price", color='blue')
plt.plot(df['Date'], df['Upper'], label="Upper Band", color='red')
plt.plot(df['Date'], df['Lower'], label="Lower Band", color='green')
plt.fill_between(df['Date'], df['Lower'], df['Upper'], color='lightgrey', alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('BTC-USD Closing Price with Volatility Bands')
plt.legend()
plt.show()
```

```python
#RSI Momentum Analysis
plt.figure(figsize=(8,5))
plt.plot(df['Date'], df['RSI'], label="RSI", color='lightblue')
plt.axhline(70, color='red', label='Overbought Threshold (70)')
plt.axhline(30, color='green', label='Oversold Threshold (30)')
plt.xlabel('Date')  
plt.ylabel('RSI Value')
plt.title('Relative Strength Index')
plt.legend()
plt.show()
```

When the Closing Price touches the Upper Band, market might be overbought; i.e. the price has risen significantly relative to the moving average.

RSI > 70 indicates the possible selling opportunity

When it touches Lower Band, it might be oversold; i.e. price has fallen significantly relative to moving average.

RSI < 30 indicates the possible buying opportunity.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of BTC-USD Features')
plt.show()
```

The features Middle band, Lower band, Upper band, Lag as well as MA7 and MA21 have a very high correlation with close.

```python
#21-Volatility Trend
plt.figure(figsize=(8,5))
plt.plot(df['Date'], df['volatility_21'], label="21-Day Volatility", color='purple')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.title('BTC-USD 21-Day Volatility Over Time')
plt.legend()
plt.show()
```

```python
#Moving Average Comparision
plt.figure(figsize=(8,5))
plt.plot(df['Date'], df['MA7'], label="7-Day MA", color='red')
plt.plot(df['Date'], df['MA21'], label="21-Day MA", color='green')
plt.xlabel('Date')
plt.ylabel('Moving Average Price (USD)')
plt.title('BTC-USD Moving Averages')
plt.legend()
plt.show()
```

As it can be observed that the red trend line indicating MA7 crosses above MA21 line mostly thus it is a bullish trend. 

If it would be visa-versa then it would be bearish trend
