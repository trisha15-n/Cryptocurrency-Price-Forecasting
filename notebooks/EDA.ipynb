```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
usd_path = Path('../data/BTC-USD_processed.csv')
inr_path = Path('../data/BTC-INR_processed.csv')
```

```python
btc_usd = pd.read_csv(usd_path, parse_dates=['Date'])
btc_inr = pd.read_csv(inr_path, parse_dates=['Date'])
```

```python
#Trend showing Closing Price 

exchange_rate = 83
btc_inr['Close_in_USD'] = btc_inr['Close'] / exchange_rate

plt.figure(figsize=(8,5))
plt.plot(btc_usd['Date'], btc_usd['Close'], label='BTC-USD', color='red')
plt.plot(btc_inr['Date'], btc_inr['Close_in_USD'], label='BTC-INR', color='blue')
plt.title('Bitcoin Closing Price Trend (USD vs INR)')
plt.xlabel('Date')
plt.ylabel('Closing Price(USD)')
plt.legend()
plt.show()
```

```python
#Daily Return Distribution
plt.figure(figsize=(8,5))

plt.subplot(1,2,1)
sns.kdeplot(btc_usd['Daily_Return'], fill=True, color='red')
plt.title('BTC-USD Daily Return Distribution')
plt.xlabel('Daily Return')

plt.subplot(1,2,2)
sns.kdeplot(btc_inr['Daily_Return'], fill=True, color='blue')
plt.title('BTC-INR Daily Return Distribution')
plt.xlabel('Daily Return')
plt.tight_layout()
plt.show()
```

```python
#Rolling Averages USD

plt.figure(figsize=(8,5))
plt.plot(btc_usd['Date'], btc_usd["MA7"], label='BTC-USD 7 Day', color='red')
plt.plot(btc_usd['Date'], btc_usd['MA21'], label='BTC-USD 21 Day', color='blue')
plt.title('BTC-USD Rolling Averages (7 & 21 Days)')
plt.xlabel('Date')
plt.ylabel('Closing Price(USD)')
plt.legend()
plt.show()
```

```python
#Rolling Averages INR

plt.figure(figsize=(8,5))
plt.plot(btc_inr['Date'], btc_inr["MA7"], label='BTC-INR 7 Day', color='red')
plt.plot(btc_inr['Date'], btc_inr['MA21'], label='BTC-INR 21 Day', color='blue')
plt.title('BTC-INR Rolling Averages (7 & 21 Days)')
plt.xlabel('Date')
plt.ylabel('Closing Price(INR)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.plot(btc_usd['Date'], btc_usd['Volatility'], color='red', label='BTC-USD')
plt.title('7 day Rolling Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility')

plt.subplot(1,2,2)
plt.plot(btc_inr['Date'], btc_inr['Volatility'], color='blue', label='BTC-INR')
plt.title('7 day Rolling Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.tight_layout()
plt.show() 
```

```python
#Correlation
merged = btc_usd.merge(btc_inr, on='Date', suffixes=('_USD', '_INR'))
corr_mat = merged[['Close_USD', "Close_INR", "Daily_Return_USD", "Daily_Return_INR"]].corr()

plt.figure(figsize=(8,5))
sns.heatmap(corr_mat, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix between BTC-USD and BTC-INR')
plt.show()
```

According to the correlation matrix, we can observe that there is not much difference between BTC-USD and BTC-INR in terms of their closing prices and daily returns. The high correlation values indicate that INR track global market trends.


There is a high volatility as it also impacted by the currency exchange rate fluctations (USD-INR)
The daily return shows large spikes and drops results in high risk-high reward
